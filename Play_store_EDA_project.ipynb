## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# converting the data type of Reviews column
df["Reviews"]=df["Reviews"].astype(int)
````

### Check Unique Values for each variable.

````python
df["Size"].unique()
````

````python
df[df["Size"]=="Varies with device"]
````

There are 13% rows which have no app size.

Other then that, the format of different app is in Mb or Kb.

The entire Size row is in object format.

As i am not taking the App Size as an important parameter for my data analysis.So i will remove the entire column....

````python
#dropping the columns current ver and android ver
df.drop(["Current Ver","Android Ver","Size"],axis=1,inplace=True)
````

````python
# checking for all the unique values that are present in Install column
df["Installs"].unique()
````

````python
# creating a new column named as Install+
df["Installs+"]=df["Installs"]
df.head()
````

````python
# writing a function to remove the commas placed between the digits of a number
def remove_comma(value):
    if "," in value:
        num=value.replace(",","")
        return num
    else:
        return value
````

````python
# applying that function to the Install+ column
df["Installs+"]=df["Installs+"].apply(remove_comma)
df.head()
````

````python
# writing a function to remove the + from the install apps
def remove_plus(value):
    if "+" in value:
        num=value.replace("+","")
        return num
    else:
        return value
````

````python
# applying that function to remove the + sign from the last of the numbers
df["Installs+"]=df["Installs+"].apply(remove_plus)
df.head()
````

````python
df["Installs+"].unique()
````

````python
# now converting the data type of the install+ column to int
df["Installs+"]=df["Installs+"].astype(int)
````

````python
# checking for the unique numbers present in the Price column
df["Price"].unique()
````

````python
# writing a function to remove the $ sign
def remove_dollar(value):
    if "$" in value:
        price=value.replace("$","")
        return price
    else:
        return value
````

````python
# applying that function to remove the $ sign from the Price column
df["Price"]=df["Price"].apply(remove_dollar)
df["Price"].unique()
````

````python
# converting the data type of the Price column
df["Price"]=df["Price"].astype(float)
df.describe()
````

````python
# cheking for the null values in the Type column
df["Type"].unique()
````

````python
df[df["Type"].isna()]
````

````python
# dropping the row which have null values
df.dropna(subset=["Type"],inplace=True)
df["Type"].unique()
````

````python
df.isnull().sum()
````

````python
# converting the Last Updated column to a standard Datatime format
df["Last Updated"]=df["Last Updated"].astype(str)
df["Last Updated"]=df["Last Updated"].apply(lambda x: x.strip())
df["Last Updated"] = pd.to_datetime(df["Last Updated"], format="%B %d, %Y")
````

````python
df.head()
````

````python
# making a new column which will only contain the year of the Last Updated column
df["year_updated"]=df["Last Updated"].dt.year
````

# Merging the two Data sets

````python
# seeing Trends in User Reviews Dataset
df2
````

As, the User Reviews Dataset have many duplicates rows, so i am grouping the entire dataset using aggregate fucntion so that i get a new Data frame which only contains the unique values...

````python
# Applying groupby function to the entire User Reviews Dataset using Aggregate
new_df = df2.groupby('App').agg({
    'Translated_Review': 'count',
    'Sentiment': lambda x: x.mode().iloc[0],
    'Sentiment_Polarity': 'mean',
    'Sentiment_Subjectivity': 'mean'
}).reset_index()

# Rename the count column
new_df.rename(columns={'Translated_Review': 'Review_Count'}, inplace=True)
new_df
````

````python
# Merging the Play store and User Reviews Dataframe for common apps only
merged_df = pd.merge(df, new_df, on='App',how='inner')
merged_df.head()
````
